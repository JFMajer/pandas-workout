```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/jakub/data/nyc_taxi_2019-01.csv", 
                 usecols=["passenger_count", "trip_distance", "total_amount"])
```

```python
result = df.groupby('passenger_count')['total_amount'].agg('mean')
```

```python
result_df = result.reset_index(name='average_total_amount')
```

```python
result_df.sort_values(by='average_total_amount', ascending=False)
```

```python
def distance_group(trip_in_miles):
    if trip_in_miles < 2:
        return "short"
    elif trip_in_miles >=2 and trip_in_miles < 10:
        return "medium"
    elif trip_in_miles >= 10:
        return "long"
```

```python
df["trip_distance_group"] = df["trip_distance"].apply(distance_group)
```

```python
df
```

```python
df["trip_distance_group"] = pd.cut(df['trip_distance'], [df['trip_distance'].min(), 2, 10, df['trip_distance'].max()], labels=['short', 'medium', 'long'])
```
